## Hough Lines with OpenCV


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image = cv2.imread('./road.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
# Resize it to 50%
image = cv2.resize(image, (0,0) , fx=0.5,fy=0.5 )
plt.imshow(image)
```

```python
# To perform hough line detection, first convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
# Apply Canny edge detector
c_edge = cv2.Canny(gray, 240,255)
plt.imshow(c_edge,cmap='gray')
```

```python
## Find continuous lines using Hough Transform
rho=1
theta = np.pi/180
threshold = 10
min_line_length = 10
max_line_gap = 5

#Find lines 
lines = cv2.HoughLinesP(c_edge, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)
line_image = np.copy(image)

for line in lines:
    for x1,y1,x2,y2 in line:
        if abs((y2-y1)/(x2-x1))>0.3:
            cv2.line(line_image, (x1,y1),(x2,y2), (255,0,0), 5)
    
plt.imshow(line_image)

```
