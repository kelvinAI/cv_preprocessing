## Blurring with low pass filters before feeding into high pass filters for edge detection

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
%matplotlib inline
```

```python
image = cv2.imread('./mri.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# Convert to grayscale first
gray_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
# Create a gaussianBlur filter. (5,5) : Kernel Size (Blur Size), 0 for stddev for it to be automatically calculated
gray_blur = cv2.GaussianBlur(gray_img, (5,5), 0)
plt.imshow(gray_blur, cmap='gray')
```

```python
# Compare with sobel edge detection filters
sobel_y = np.array([[-1,-2,-1],
                   [0,0,0],
                   [1,2,1]])

filtered_normal = cv2.filter2D(gray_img, -1 , sobel_y)
filtered_blurred = cv2.filter2D(gray_blur, -1, sobel_y)

fig, ax = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(20,10))
ax[0].set_title("Original Version")
ax[0].imshow(filtered_normal, cmap='gray')
ax[1].set_title("Blurred version")
ax[1].imshow(filtered_blurred, cmap='gray')
```
