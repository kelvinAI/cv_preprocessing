```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
%matplotlib inline
```

```python
image = cv2.imread('city.jpg')
plt.imshow(image)
```

```python
#Image is BGR, convert them to RGB
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
# Convert image to grayscale to enhance edge intensity
gray = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

## Create a custom Kernel

```python
sobel_x = np.array([[-1,0,1],
                    [-2,0,1],
                    [-1,0,1]])

#Perform convolutions using filter2D
filtered_image= cv2.filter2D(gray, -1, sobel_x)
plt.imshow(filtered_image, cmap='gray')
```

```python
# Create a binary image from the filtered image
retval, binary_image = cv2.threshold(filtered_image, 100, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap='gray')

#Vertical Lines are enhanced
```
