```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 
%matplotlib inline
```

```python
img = cv2.imread('./bird.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# First convert it to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
# Implement Canny edge detection
lower_bound= 120
upper_bound = 240

edges = cv2.Canny(gray, lower_bound, upper_bound)
plt.imshow(edges, cmap='gray')
```

```python
wide = cv2.Canny(gray, 80,150)
tight = cv2.Canny(gray, 160,240)
f, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,10))
ax[0].set_title("Wide (Low threshold)")
ax[0].imshow(wide,cmap='gray')
ax[1].set_title("Tight (High threshold)")
ax[1].imshow(tight, cmap='gray')
```
